```python
# Import the necessary modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Bring each CSV into a separate data frame
wwe_2013_df = pd.read_csv("Resources/WWE-Data-2013.csv")
wwe_2014_df = pd.read_csv("Resources/WWE-Data-2014.csv")
wwe_2015_df = pd.read_csv("Resources/WWE-Data-2015.csv")
wwe_2016_df = pd.read_csv("Resources/WWE-Data-2016.csv")
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
wwe_2013and2014_df = pd.merge(wwe_2013_df, wwe_2014_df, how="left", on=["Wrestler", "Wrestler"])
wwe_2013and2014_df
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"
wwe_2013and2014_df.rename(columns = {'Wins_x':'2013 Wins'}, inplace = True)
wwe_2013and2014_df.rename(columns = {'Losses_x':'2013 Losses'}, inplace = True)
wwe_2013and2014_df.rename(columns = {'Draws_x':'2013 Draws'}, inplace = True) 
# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"

wwe_2013and2014_df.rename(columns = {'Wins_y':'2014 Wins'}, inplace = True)
wwe_2013and2014_df.rename(columns = {'Losses_y':'2014 Losses'}, inplace = True)
wwe_2013and2014_df.rename(columns = {'Draws_y':'2014 Draws'}, inplace = True) 
```
